### The Dot Product

If we have two vectors (say, ***x*** and ***y***) with the same length *n*, we can calculate the dot product between them. This is annotated several different ways, including the following:

* $x \cdot y$
* $x^Ty$
* $\langle x,y \rangle$

Regardless which notation you use (I prefer the first), the calculation is the same; we calculate products in an element-wise fashion and then sum reductively across the products to a scalar value. That is, $x \cdot y = \sum_{i=1}^{n} x_i y_i$

The dot product is ubiquitous in deep learning: It is performed at every artificial neuron in a deep neural network, which may be made up of millions (or orders of magnitude more) of these neurons.

### Frobenius Norm

The **Frobenius norm** of a matrix is a way to measure its overall size.

For a matrix $$ A = [a_{ij}] $$, it is defined as

$$ |A|_F = \sqrt{\sum_{i,j} a_{ij}^2} $$


In words: **square all entries, add them up, then take the square root**.
It’s essentially the Euclidean length of the matrix when viewed as a long vector.

### Symmetric Matrices

**Symmetric Matrix**: A square matrix that equals its transpose, meaning $A = A^T$. For any element $a_{ij} = a_{ji}$ - elements are mirrored across the main diagonal.

Example: $\begin{bmatrix} 1 & 2 & 3 \\ 2 & 5 & 4 \\ 3 & 4 & 6 \end{bmatrix}$

**Identity Matrix**: A square matrix with 1s on the main diagonal and 0s elsewhere, denoted as $I$ or $I_n$ for an $n \times n$ matrix. It's the multiplicative identity for matrices: $AI = IA = A$ for any compatible matrix $A$.

Example: $I_3 = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$

Note: Every identity matrix is symmetric.

### Matrix Inversion

Matrix inversion is the process of finding a matrix $A^{-1}$ such that $A \cdot A^{-1} = A^{-1} \cdot A = I$, where $I$ is the identity matrix. Only square matrices with non-zero determinants are invertible.

**Key uses:**
- **Solving linear systems**: Instead of $Ax = b$, compute $x = A^{-1}b$
- **Regression analysis**: Computing least squares solutions like $(X^TX)^{-1}X^Ty$ in linear regression
- **Computer graphics**: Reversing transformations (rotation, scaling, translation)
- **Control theory**: System analysis and feedback design
- **Cryptography**: Encoding and decoding operations

**Important notes:**
- Matrix inversion is computationally expensive ($O(n^3)$)
- In practice, solving $Ax = b$ directly is more efficient than computing $A^{-1}$ first
- Numerical stability issues can arise with near-singular matrices (determinant ≈ 0)
- Alternative methods like LU decomposition or iterative solvers are often preferred for large systems

**Using identity matrix technique: (Square Matrices)**

Starting from $Xw = y$:

1. Multiply both sides by $X^{-1}$:
   $$X^{-1}Xw = X^{-1}y$$

2. Apply identity property ($X^{-1}X = I$):
   $$Iw = X^{-1}y$$

3. Since $Iw = w$:
   $$w = X^{-1}y$$

**For rectangular X (overdetermined system):**

1. Start: $Xw = y$

2. Multiply both sides by $X^T$:
   $$X^TXw = X^Ty$$

3. Multiply both sides by $(X^TX)^{-1}$:
   $$(X^TX)^{-1}X^TXw = (X^TX)^{-1}X^Ty$$

4. Apply identity: $(X^TX)^{-1}X^TX = I$:
   $$Iw = (X^TX)^{-1}X^Ty$$

5. Therefore:
   $$w = (X^TX)^{-1}X^Ty$$

**Example**:

Example of house regression system of equations.
$$\begin{align}
4b + 2c &= 4 \\
-5b - 3c &= -7
\end{align}$$

Or in matrix form:

$$\begin{bmatrix} 4 & 2 \\ -5 & -3 \end{bmatrix} \begin{bmatrix} b \\ c \end{bmatrix} = \begin{bmatrix} 4 \\ -7 \end{bmatrix}$$

$$\mathbf{w} = \begin{bmatrix} w_1 \\ w_2 \end{bmatrix} = \begin{bmatrix} b \\ c \end{bmatrix} = X^{-1}\mathbf{y}$$

Where:

$$X = \begin{bmatrix} 4 & 2 \\ -5 & -3 \end{bmatrix}, \quad \mathbf{y} = \begin{bmatrix} 4 \\ -7 \end{bmatrix}$$

```python
X = np.array([[4, 2], [-5, -3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

### Matrix Inversion Where No Solution

**Square system:** Equal number of equations and unknowns ($m = n$). The matrix $X$ is square. Typically has **one unique solution** if the matrix is invertible (non-singular). Solve directly as $w = X^{-1}y$.

**Overdetermined system:** More equations than unknowns ($m > n$). The matrix $X$ has more rows than columns. Typically no exact solution exists, so we find the **least squares approximation** that minimizes error. Common in regression with many data points.

**Underdetermined system:** Fewer equations than unknowns ($m < n$). The matrix $X$ has more columns than rows. Typically has **infinitely many solutions**. We often find the minimum norm solution or add regularization to select one unique solution.

**Examples:**
- Square: 3 equations, 3 unknowns → $X$ is $3 \times 3$
- Overdetermined: 100 data points, 3 features → $X$ is $100 \times 3$
- Underdetermined: 3 data points, 100 features → $X$ is $3 \times 100$

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Square System: 2x2 (unique solution)
ax1 = axes[0]
x = np.linspace(-1, 3, 100)
# 2x + y = 3  =>  y = 3 - 2x
# x - y = 0   =>  y = x
y1_square = 3 - 2*x
y2_square = x
ax1.plot(x, y1_square, 'b-', label='2x + y = 3', linewidth=2)
ax1.plot(x, y2_square, 'r-', label='x - y = 0', linewidth=2)
ax1.plot(1, 1, 'go', markersize=10, label='Unique solution (1, 1)')
ax1.set_xlim(-0.5, 2.5)
ax1.set_ylim(-0.5, 3.5)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Square System (m = n)\n2 equations, 2 unknowns\nUnique Solution', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)

# Overdetermined System: 3 equations, 2 unknowns (no exact solution)
ax2 = axes[1]
x = np.linspace(-1, 3, 100)
# Three lines that don't intersect at a single point
# x + y = 2    =>  y = 2 - x
# 2x + y = 3   =>  y = 3 - 2x
# x + 2y = 3   =>  y = (3 - x)/2
y1_over = 2 - x
y2_over = 3 - 2*x
y3_over = (3 - x)/2
ax2.plot(x, y1_over, 'b-', label='x + y = 2', linewidth=2)
ax2.plot(x, y2_over, 'r-', label='2x + y = 3', linewidth=2)
ax2.plot(x, y3_over, 'g-', label='x + 2y = 3', linewidth=2)
ax2.plot(1, 1, 'mo', markersize=10, label='Least squares solution')
ax2.set_xlim(-0.5, 2.5)
ax2.set_ylim(-0.5, 3.5)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Overdetermined (m > n)\n3 equations, 2 unknowns\nNo Exact Solution', fontsize=12, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.axvline(x=0, color='k', linewidth=0.5)

# Underdetermined System: 1 equation, 2 unknowns (infinite solutions)
ax3 = axes[2]
x = np.linspace(-1, 3, 100)
# x + y = 2  =>  y = 2 - x (infinite points on this line)
y_under = 2 - x
ax3.plot(x, y_under, 'b-', label='x + y = 2', linewidth=3)
ax3.scatter([0, 1, 2], [2, 1, 0], color='orange', s=100, zorder=5, label='Some solutions', alpha=0.7)
ax3.set_xlim(-0.5, 2.5)
ax3.set_ylim(-0.5, 3.5)
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('Underdetermined (m < n)\n1 equation, 2 unknowns\nInfinite Solutions', fontsize=12, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)
ax3.axhline(y=0, color='k', linewidth=0.5)
ax3.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()
```

### Diagonal Matrices
Nonzero elements along the main diagonal, zero everywhere else. Identity matrix and an example.
-  diag(x) where x is the diagonal elements.


#### Computational Efficiency of Diagonal Matrices

**Matrix Multiplication:**
- Standard matrix multiplication: $O(n^3)$ operations
- Diagonal matrix multiplication: $O(n)$ operations
- Simply multiply each element by the corresponding diagonal value

**Matrix Inversion:**
- Standard inversion: $O(n^3)$ operations
- Diagonal matrix inversion: $O(n)$ operations
- Invert by taking reciprocal of each diagonal element: $D^{-1}_{ii} = \frac{1}{D_{ii}}$

**Example:**
$$D = \begin{bmatrix} 2 & 0 & 0 \\ 0 & 3 & 0 \\ 0 & 0 & 5 \end{bmatrix} \Rightarrow D^{-1} = \begin{bmatrix} 0.5 & 0 & 0 \\ 0 & 0.33 & 0 \\ 0 & 0 & 0.2 \end{bmatrix}$$

This makes diagonal matrices **extremely efficient** for large-scale computations in machine learning and scientific computing.

#### Non-Square Diagonal Matrices

Diagonal matrices can also be **rectangular** ($m \times n$ where $m \neq n$):

**Tall diagonal matrix** ($m > n$):
$$D = \begin{bmatrix} 2 & 0 & 0 \\ 0 & 3 & 0 \\ 0 & 0 & 5 \\ 0 & 0 & 0 \end{bmatrix}_{4 \times 3}$$

**Wide diagonal matrix** ($m < n$):
$$D = \begin{bmatrix} 2 & 0 & 0 & 0 \\ 0 & 3 & 0 & 0 \\ 0 & 0 & 5 & 0 \end{bmatrix}_{3 \times 4}$$

**Key properties:**
- **Cannot be inverted** (not square), but have a **pseudoinverse**
- **Multiplication is still $O(\min(m,n))$** - extremely efficient
- **Common in dimensionality reduction** (e.g., SVD, PCA) where they scale and truncate dimensions
- **Pseudoinverse**: Transpose and invert non-zero diagonal elements

**Example pseudoinverse:**
$$D = \begin{bmatrix} 2 & 0 \\ 0 & 3 \\ 0 & 0 \end{bmatrix} \Rightarrow D^+ = \begin{bmatrix} 0.5 & 0 & 0 \\ 0 & 0.33 & 0 \end{bmatrix}$$

Used extensively in **singular value decomposition (SVD)** for matrix factorization and data compression.

### Orthogonal Matrices

**Definition:** A matrix $Q$ is orthogonal if its columns are **orthonormal** - mutually orthogonal (perpendicular) unit vectors.

**Key property:**
$$Q^TQ = QQ^T = I$$

This means $Q^T = Q^{-1}$ - the transpose equals the inverse.

**Why the inverse equals the transpose:**
1. For an orthogonal matrix $A$, the orthonormal columns give us: $A^TA = I$
2. We know that for any matrix: $A^{-1}A = I$ (definition of inverse)
3. Comparing these: $A^TA = I$ and $A^{-1}A = I$
4. Both equations show something times $A$ equals $I$
5. Therefore: $A^T = A^{-1}$

In plain words: Since $A^T$ does the same job as $A^{-1}$ (both give the identity when multiplied with $A$), they must be the same thing!

**Properties:**
- **Preserves lengths**: $\|Qx\| = \|x\|$ (no stretching or shrinking)
- **Preserves angles**: Dot products unchanged
- **Computationally efficient**: Inverse is just the transpose (no expensive matrix inversion)
- **Numerically stable**: Avoids numerical errors in calculations

**Applications:**
- Rotation and reflection transformations in 3D graphics
- QR decomposition for solving linear systems
- Principal Component Analysis (PCA)
- Orthonormal basis construction

**Example:**
$$Q = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \quad \Rightarrow \quad Q^T = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} = Q^{-1}$$

**Steps to Prove a Matrix is Orthogonal:**

**Method 1: Check orthonormality of columns**
1. Extract each column of the matrix
2. **Verify orthogonality**: Check that dot product of any two different columns equals 0
3. **Verify unit length**: Check that each column has norm (length) equal to 1
4. If both conditions hold for all columns, the matrix is orthogonal

**Method 2: Verify the key property**
1. Compute $A^TA$ (transpose times original)
2. Check if $A^TA = I$ (identity matrix)
3. If true, the matrix is orthogonal

**Method 3: Check both directions**
1. Compute both $A^TA$ and $AA^T$
2. Check if both equal the identity matrix: $A^TA = AA^T = I$
3. This confirms both columns AND rows are orthonormal

Note: Method 2 is usually the quickest computational approach, while Method 1 provides more intuitive understanding.
